```python
import numpy as np
from scipy.stats import norm
import matplotlib.pyplot as plt
  
# Gere alguns dados para isso
# Demonstração.
data = np.random.normal(estrutura_salarial_vca.mean())
  
# Ajuste uma distribuição normal para os dados
# média e desvio padrão
mu, std = norm.fit(data) 
  
# Plotar o histograma.
plt.hist(data, bins=25, density=True, alpha=0.6, color='b')
  
# Plotar o PDF.
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, mu, std)
  
plt.plot(x, p, 'r', linewidth=2)
title = "Gráfico 8: Distribuição salarial entre os subsetores da economia do mercado de trabalho formal de Vitória da Conquista no ano de 2015 considerando apenas trabalhadores com vínculo CLT U/PJ IND e carga horária entre 40 e 44 horas semanais".format(mu, std)
plt.title(title)
  
plt.show()
```
